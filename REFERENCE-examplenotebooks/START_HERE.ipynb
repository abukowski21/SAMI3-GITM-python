# How to use a few of the programs here


## Using the libraries...
------

I've spent a ton of time developing a softwarte stack to deal with the model outputs. Most of the regridding options can be called with the postprocessmodelresults.py program. From the command line you can see what the options are with `python PostprocessModelResults.py -h`

Not all SAMI regridding methods are exposed there. You will need to use `python RegridSami.py -h` to see those options.

> the `-h` flag is to ask the program for help... Most of the `.py` files have some documentation built in to get help on whatt is exposed on the command line.

## Importing the functions into a jupyter notebook
------

To import anything, you have two options. You can pick either, but it will depend where your jupyter notebook is in relaion to the utility you are trying to import.

### Option 1:
If your jupyter notebook is in the root directory of this package, you can just run the code below:

`from utility_programs.read_routines import SAMI`
(Or whatever you want to import)

> To make sure this will work (if you don't want to just go for it), running `!ls` in a code cell should show you if you're in the right folder.




### Option 2:
If you're in a different directory than utility_programs/ (or RegridSami.py pr whatever...)

You need to tell python the correct path to look for what you're trying to import. Do this with:

```
import sys
sys.path.append('[PATH_TO_CORRECT_DIRECTORY]) 

```

> *swap [PATH_TO_CORRECT_DIRECTORY] with the path to the package root directory*
> For you, you can probably use `sys.path.append('~/SAMI3-GITM-python/')` Just make 

Then you can import with:

```
from utility_programs import filters
```

And use away!!
