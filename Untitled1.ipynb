```python
def transform_datasets_by_model_v2(in_glob, out_name):
    
    if 'SAMI' in in_glob:
        ds = xr.open_mfdataset(in_glob, concat_dim='time', combine='nested',
                               parallel=True, engine='h5netcdf')
    else:
        ds = xr.open_mfdataset(in_glob, 
                               parallel=True, engine='h5netcdf')
        
    ds.to_netcdf('/glade/u/home/abukowski/scratch/postprocessed/' + out_name, encoding={'time':{'dtype':'float'}})
    
    return ds
```

```python
in_globs = []
out_names = []

for i in mapping.keys():
    if len(glob.glob(os.path.join(i, '*GITM*'))) > 1:
        in_globs.append(os.path.join(i, '*GITM*'))
        out_names.append(mapping[i]+'_GITM.nc')
    
    if len(glob.glob(os.path.join(i, '*RAW*'))) > 1:
        in_globs.append(os.path.join(i, '*RAW*'))
        out_names.append(mapping[i]+'_SAMI_RAW.nc')
        
    if len(glob.glob(os.path.join(i, '*REGRID*'))) > 1:
        in_globs.append(os.path.join(i, '*REGRID*'))
        out_names.append(mapping[i]+'_SAMI_REGRID.nc')
```

```python
def transform_datasets(in_dir, out_name):
    do_sami_raw = False
    do_sami_regrid=False
    do_gitm = False
    if len(glob.glob(os.path.join(in_dir,'*GITM*'))) != 0:
        do_gitm = True
    if len(glob.glob(os.path.join(in_dir,'*SAMI_REGRID*'))) != 0:
        do_sami_regrid = True
    if len(glob.glob(os.path.join(in_dir,'*SAMI_RAW*'))) != 0:
        do_sami_raw = True
        
    if do_gitm:
        ds = xr.open_mfdataset(os.path.join(in_dir,'*GITM*'), concat_dim='time', combine='nested',
                               parallel=True, engine='h5netcdf', chunks={'time':1000})
        
        ds.to_netcdf('/glade/u/home/abukowski/scratch/postprocessed/' + out_name + '_GITM.nc',
                     engine='h5netcdf', encoding={'time':{"dtype": "float"}})
    
    if do_sami_raw:
        ds = xr.open_mfdataset(os.path.join(in_dir,'*SAMI_RAW*'), concat_dim='time', combine='nested',
                               parallel=True, engine='h5netcdf', chunks={'time':1000})
        
        ds.to_netcdf('/glade/u/home/abukowski/scratch/postprocessed/' + out_name + '_SAMI_RAW.nc',
                     engine='h5netcdf', encoding={'time':{"dtype": "float"}})
    
    if do_sami_regrid:
        ds = xr.open_mfdataset(os.path.join(in_dir,'*SAMI_REGRID*'), concat_dim='time', combine='nested',
                               parallel=True, engine='h5netcdf', chunks={'time':1000})
        
        ds.to_netcdf('/glade/u/home/abukowski/scratch/postprocessed/' + out_name + '_SAMI_REGRID.nc',
                     engine='h5netcdf', encoding={'time':{"dtype": "float"}})
    
    print("done with moving %i filetypes from %s to new folder" %(np.sum([do_gitm, do_sami_raw, do_sami_regrid]), in_dir))
```

```python
futs = []
for i in range(len(list(mapping.keys()))):
    futs.append(client.submit(transform_datasets, list(mapping.keys())[i], list(mapping.values())[i]))
```
